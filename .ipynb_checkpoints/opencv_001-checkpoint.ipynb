(1) Reading Images

```python
import numpy as np
```

When an image is read it results in (x,y,z) form

Where x = pixel Height 

y = pixel Width

z= Color Channels(Red,Green and Blue)


```python
from PIL import Image
```

```python
pic = Image.open('multiverse.jpg')
```

```python
type(pic)
```

```python
print(pic.format, pic.size, pic.mode)
```

```python
pic.show()
```

```python
np_pic = np.asanyarray(pic)
```

```python
type(np_pic)
```

```python
np_pic.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(np_pic)
```

```python
plt.imshow(np_pic[:,:,0])
```

the above is Red image as 0 corresponds to Red

```python
plt.imshow(np_pic[:,:,1]) #this image corresponds to green
```

```python
plt.imshow(np_pic[:,:,2]) # this image corresponds to blue
```

```python
plt.imshow(np_pic[:,:,0],cmap='gray')
```

```python
plt.imshow(np_pic[:,:,1],cmap='gray')
```

```python
plt.imshow(np_pic[:,:,2],cmap='gray')
```
