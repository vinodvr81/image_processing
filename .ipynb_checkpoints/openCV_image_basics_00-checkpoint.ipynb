```python
cake_img = cv2.imread('elaicheesecake.png')
```

```python
new_cake_flip_img = cv2.flip(cake_img,0) #Along central x axis
plt.imshow(new_cake_flip_img)
```

```python
plt.imshow(cake_img)
```

```python
new_cake_another_flip = cv2.flip(cake_img,1) # along central y axis
plt.imshow(new_cake_another_flip)
```

```python
both_flip_img = cv2.flip(cake_img,-1)  # flip on both axis
plt.imshow(both_flip_img)
```

```python
rgb_cake = cv2.cvtColor(both_flip_img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb_cake)
```

```python
cv2.imwrite('my_new_flip_cake_both_axis_picture.jpg',rgb_cake)  #saving images
```
