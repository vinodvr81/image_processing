```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
black_image = np.zeros(shape=(256,256,3),dtype=np.int16)
```

```python
plt.imshow(black_image)
```

```python
black_image.shape
```

```python
black_image.size
```

Circles

```python
cv2.circle(img=black_image, center=(125,125), radius=30, color=(0,255,255), thickness=2)
plt.imshow(black_image)
```

```python
cv2.circle(img=black_image, center=(200,200), radius=20, color=(255,255,0), thickness=-1)
plt.imshow(black_image)
```

```python
cv2.rectangle(black_image,pt1=(50,10),pt2=(75,35),color=(0,255,0), thickness=-1)
plt.imshow(black_image)
```

Lines

```python
cv2.line(black_image,pt1=(0,0),pt2=(256,256),color=(200, 255, 255),thickness=1)
plt.imshow(black_image)
```

```python
font = cv2.FONT_ITALIC
cv2.putText(black_image,text='Hello Vinod Vukkalam',org=(10,245), fontFace=font,fontScale= 0.5,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(black_image)
```
